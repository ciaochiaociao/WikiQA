```python
def in_ranges(ix, ranges):
    for tok_b, tok_e in ranges:
        if tok_b <= ix < tok_e:
            return True
    return False

in_ranges(5, [(2,4), (6,8)])
```

```python
from ansi.color import fg, bg
import re

def highlight(text, match, output=False):
    highlighted = re.sub(r'(' + match + ')', fg.red(r'\1'), text)
    if output:
        print(highlighted)
    else:
        return highlighted
```
